```python
# import dependencies
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None 
import folium
from folium import plugins
```

```python
# import csv

ANTARCTICA_CSV = pd.read_csv("./Data/Antarctica_Hotpoints.csv")
```

```python
# 2000-2021 data
GENERAL_POINTS_COOR = ANTARCTICA_CSV[["latitude","longitude"]]

GENERAL_POINTS_COOR.latitude = GENERAL_POINTS_COOR.latitude.astype(float)
GENERAL_POINTS_COOR.longitude = GENERAL_POINTS_COOR.longitude.astype(float)


GENERAL_HEAT = list(zip(GENERAL_POINTS_COOR.latitude,GENERAL_POINTS_COOR.longitude))
```

```python
# 2021 data
ANTARCTICA_2021 = ANTARCTICA_CSV[-529:-1]

ANTARCTICA_2021.reset_index(drop=True,inplace=True)

ANTARCTICA_2021
```

## HOTPOINTS 2000-2021

```python
MAP_FUNCTION = folium.Map(location=[-30.52523,150.157791],
                  tiles='Stamen Terrain',
                  zoom_start=2.0)

plugins.MarkerCluster(GENERAL_POINTS_COOR).add_to(MAP_FUNCTION)

MAP_FUNCTION
```

```python
MAP_FUNCTION = folium.Map(location=[-30.52523,150.157791],
                  tiles='Stamen Terrain',
                  zoom_start=2.0)

plugins.HeatMap(GENERAL_HEAT).add_to(MAP_FUNCTION)

MAP_FUNCTION
```
